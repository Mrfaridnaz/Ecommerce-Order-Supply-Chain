## Top 20 Cities based on sale

```python
# Group by 'category' and calculate the sum of 'sales'
sales_sum = Data1.groupby('customer_city')['payment_value'].sum()
sales_sum
```

```python
import matplotlib.pyplot as plt

# Grouped sales sum by customer_city
sales_sum = Data1.groupby('customer_city')['payment_value'].sum()

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(sales_sum, bins=20, color='skyblue', edgecolor='black')

# Adding title and labels
plt.title('Distribution of Total Sales by Customer City')
plt.xlabel('Total Sales')
plt.ylabel('Frequency')
plt.grid(axis='y')

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Find the top 10 sales
top_10_sales = sales_sum.nlargest(20)
top_10_sales
```

```python
top_10_sales.mean()
```

```python
import matplotlib.pyplot as plt
```

```python
# Exclude the first 2 data points
top_10_sales_exclude_first_2 = top_10_sales.iloc[2:]
top_10_sales_exclude_first_2
```

```python
# Plotting
plt.figure(figsize=(10, 6))
ax = top_10_sales_exclude_first_2.plot(kind='bar', color='skyblue')

# Adding values on top of each bar in millions
for p in ax.patches:
    ax.annotate(f'{p.get_height() / 1e6:.2f}M', 
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom')

# Adding title and labels
plt.title('20 Top Sales by Customer City (Excluding First 2)')
plt.xlabel('Customer City')
plt.ylabel('Total Sales (in Millions)')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'data' is a DataFrame with 'customer_state' and 'customer_id'

# Group by state to get customer distribution
state_distribution = data.groupby('customer_state')['customer_id'].count().reset_index()

# Rename the columns for easier reference
state_distribution.columns = ['State', 'Customer Count']

# Sort the values to find the top state by customer count
state_distribution = state_distribution.sort_values(by='Customer Count', ascending=False)

# Skip the top state by excluding the first row
state_distribution = state_distribution.iloc[1:]

# --- Bar Plot: Customer Distribution by State (excluding top state) ---
plt.figure(figsize=(10, 6))
sns.barplot(x='State', y='Customer Count', data=state_distribution, palette='viridis')
plt.xticks(rotation=90)
plt.title('Customer Distribution by State (Excluding Top State)')
plt.xlabel('State')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.show()
```

```python
# Required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'data' is a DataFrame with 'customer_city' and 'customer_id'

# Group by city to get customer distribution
city_distribution = data.groupby('customer_city')['customer_id'].count().reset_index()

# Rename the columns for easier reference
city_distribution.columns = ['City', 'Customer Count']

# Sort the values to get the top 10 cities by customer count
city_distribution = city_distribution.sort_values(by='Customer Count', ascending=False)

# Remove the top city by excluding the first row
top_9_cities = city_distribution.iloc[1:10]  # Skipping the top city and taking the next 9

# --- Pie Plot: Customer Distribution by Remaining Top 9 Cities ---
plt.figure(figsize=(7, 7))
plt.pie(top_9_cities['Customer Count'], labels=top_9_cities['City'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(top_9_cities)))
plt.title('Customer Distribution by Remaining Top 9 Cities')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
# Required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'data' is a DataFrame with 'product_category_name', 'customer_state'

# Group by state and product category to get the purchase count
category_state_distribution = data.groupby(['customer_state', 'product_category_name'])['customer_id'].count().reset_index()

# Rename columns for easier reference
category_state_distribution.columns = ['State', 'Product Category', 'Purchase Count']

# Sort values by state and purchase count
category_state_distribution = category_state_distribution.sort_values(by=['State', 'Purchase Count'], ascending=[True, False])

# Get the top product categories for each state
top_categories_per_state = category_state_distribution.groupby('State').head(10)  # Adjust number as needed

# --- Bar Plot: Top Product Categories by State ---
plt.figure(figsize=(12, 8))
sns.barplot(x='Purchase Count', y='Product Category', hue='State', data=top_categories_per_state, palette='viridis')
plt.title('Top Product Categories Purchased by State')
plt.xlabel('Number of Purchases')
plt.ylabel('Product Category')
plt.tight_layout()
plt.show()
```

```python
# Required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'data' is a DataFrame with 'product_category_name', 'customer_city'

# Group by city and product category to get the purchase count
category_city_distribution = data.groupby(['customer_city', 'product_category_name'])['customer_id'].count().reset_index()

# Rename columns for easier reference
category_city_distribution.columns = ['City', 'Product Category', 'Purchase Count']

# Sort values by city and purchase count
category_city_distribution = category_city_distribution.sort_values(by=['City', 'Purchase Count'], ascending=[True, False])

# Get the top product categories for each city
top_categories_per_city = category_city_distribution.groupby('City').head(10)  # Adjust number as needed

# --- Bar Plot: Top Product Categories by City ---
plt.figure(figsize=(12, 8))
sns.barplot(x='Purchase Count', y='Product Category', hue='City', data=top_categories_per_city, palette='viridis')
plt.title('Top Product Categories Purchased by City')
plt.xlabel('Number of Purchases')
plt.ylabel('Product Category')
plt.tight_layout()
plt.show()
```
